```python
import pandas as pd
import json, requests
```

```python
response = requests.get('https://jsonplaceholder.typicode.com/posts')
data = json.loads(response.text)

print(data[0]['title'])
```

```python
# extract relevant fields

user_id = []
id = []
title = []
body = []

for i in data:
    user_id.append(i['userId'])
    id.append(i['id'])
    title.append(i['title'])
    body.append(i['body'])
    
user_id
```

```python
# store in a dictionary

data_dict = {
    "user_id": user_id,
    "id": id,
    "title": title,
    "body": body
}

data_dict
```

```python
df = pd.DataFrame(data_dict, columns=["user_id", "id", "title", "body"])
df.head()
```

```python
def get_paginated(url, offset):
    try:
        response_data = []
        limit = offset
        initial_offset = offset
        initial_url = f'{url}?offset=0&limit={limit}'
        print('Sending request to: ' + initial_url)
        initial_response = requests.get(initial_url)
        initial_data = initial_response.json()
        #print(initial_data)
        response_data.extend(initial_data['results'])
        while True:
            paginated_url = f'{url}?offset={offset}&limit={limit}'
            print('Sending request to: ' + paginated_url)
            response_1 = requests.get(paginated_url)
            data = response_1.json()
            #print(data)
            if len(data['results']) == 0:
                break
            else:
                response_data.extend(data['results'])
                offset += initial_offset
        return response_data
    except Exception as e:
        print(e)
```

```python
def save_to_df(data_response):
    name = []
    url = []

    for i in data_response:
        name.append(i['name'])
        url.append(i['url'])
        
    data_dict = {
    'name': name,
    'url': url
    }
    
    df = pd.DataFrame(data_dict, columns=['name','url'])
    return df
    
```

```python
output = get_paginated('http://pokeapi.co/api/v2/pokemon', 100)
df = save_to_df(output)
```
